# **Manipulating dataframe**


What were the 5 most popular baby names in 2014 in US?

*   Slice out rows for 2014
*   Sort the rows in descending order by the count
*   Retrieve the first five rows
---
**US Baby Names at KAGGLE** --> [Download](https://www.kaggle.com/datasets/kaggle/us-baby-names)


```python
import pandas as pd
```

```python
path = "/content/us_baby_names.csv"
us_babies = pd.read_csv(path)
```

```python
us_babies
```

```python
us_babies["Year"]
```

```python
us_babies["Year"]==2014
```

```python
us_babies_2014 = us_babies.loc[us_babies['Year']==2014, :]
```

```python
us_babies_2014
```

```python
sort_us_2014 = us_babies_2014.sort_values('Count',ascending=False)
```

```python
sort_us_2014
```

```python
sort_us_2014.iloc[0:5]
```

**So, Emma, Olivia, Noah, Sophia, and Liam is the 5 most popular baby names in 2014 in US**
